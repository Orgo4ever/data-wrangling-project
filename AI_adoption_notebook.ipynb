```python
import requests
import pandas as pd
from io import StringIO

# 1. Base API endpoint
url = (
    "https://ec.europa.eu/eurostat/api/dissemination/sdmx/3.0/data/"
    "dataflow/ESTAT/ISOC_EB_AI/1.0"
)

# 2. Filters sent with the request
params = {
    "format": "csvdata",
    "formatVersion": "2.0",
    "compress": "false"

}

# 3. Send GET request
response = requests.get(url, params=params)

# 4. Check whether it worked
print(response.status_code)

# 5. Turn returned CSV text into a DataFrame
ai_adoption_df = pd.read_csv(StringIO(response.text))

# 6. Inspect the result
ai_adoption_df.head()
```

## Cleaned AI Adoption Dataset

This dataset was collected from the Eurostat API and measures the percentage of enterprises with 10+ employees using at least one AI technology.

Cleaning steps:

* Filtered to annual observations (`freq = A`)
* Kept enterprises with 10+ employees (`size_emp = GE10`)
* Kept the “uses at least one AI technology” indicator (`E_AI_TANY`)
* Kept percentage of enterprises (`PC_ENT`)
* Removed technical metadata, regional aggregates, and rows with missing values
* Confirmed one observation per country-year

Final columns:

* `geo`: country code
* `time_period`: reporting year
* `obs_value`: AI adoption percentage
