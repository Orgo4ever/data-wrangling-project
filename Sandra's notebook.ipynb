```python
import pandas as pd

url = "https://raw.githubusercontent.com/Orgo4ever/data-wrangling-project/main/Data/Raw%20data/eurostat_renewable_energy_share.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis=1, how="all")
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
df = df.drop(columns=[
    "STRUCTURE",
    "STRUCTURE_ID",
    "STRUCTURE_NAME",
    "freq",
    "Time frequency",
    "nrg_bal",
    "Energy balance",
    "unit",
    "Unit of measure"
])
```

```python
df = df.rename(columns={
    "Geopolitical entity (reporting)": "country",
    "TIME_PERIOD": "year",
    "OBS_VALUE": "renewable_energy_share"
})
```
